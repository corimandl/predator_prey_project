```python
import jax.numpy as jnp
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_theme(style="darkgrid")
palette = "viridis"
sns.set_palette(palette)

DATA_PATH = "./data/sheep_wolf_data2/"
SAVE_PATH = "./plots/"
```

```python
mean_energy_sheep = jnp.load(DATA_PATH + 'mean_sheep_fitness_list.npy')
mean_energy_wolves = jnp.load(DATA_PATH + 'mean_wolf_fitness_list.npy')

fig, ax = plt.subplots(figsize=(10, 6))
plt.xlabel("Timestep", fontsize=14)
ax.tick_params(axis='both', which='major', labelsize=12)

plt.plot(mean_energy_sheep, label="Sheep", linestyle='-')
plt.plot(mean_energy_wolves, label="Wolves", linestyle='-')

plt.legend(loc='upper right', fontsize=18)
plt.savefig(SAVE_PATH + 'energy_plot1.png', bbox_inches='tight')
plt.show()
```
