```python
import jax.numpy as jnp
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_theme(style="darkgrid")
palette = "viridis"
sns.set_palette(palette)
```

```python
# seed 11
mean_fitness_data = jnp.load("./params/seed_11/mean_fitness_list.npy")
min_fitness_data = jnp.load("./params/seed_11/min_fitness_list.npy")
min_fitness_data = jnp.convolve(min_fitness_data, jnp.ones(10)/10, mode='valid')
max_fitness_data = jnp.load("./params/seed_11/max_fitness_list.npy")
max_fitness_data = jnp.convolve(max_fitness_data, jnp.ones(10)/10, mode='valid')


fig,ax = plt.subplots(figsize=(10,5))
plt.xlabel("Generation", fontsize=15)
ax.tick_params(axis='both', which='major', labelsize=12)
plt.plot(mean_fitness_data, label="Mean fitness", linestyle='-')

plt.plot(min_fitness_data, label="Min fitness", linestyle='--')
plt.plot(max_fitness_data, label="Max fitness", linestyle=':')

plt.legend(loc='upper right', fontsize=18)
plt.savefig('./figs/fitness_seed_11.png', bbox_inches='tight')
plt.show()
```
